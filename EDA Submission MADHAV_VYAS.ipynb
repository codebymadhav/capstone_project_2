#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code

# Select a subset of relevant numerical columns for the pair plot
subset = df[['lead_time', 'adr', 'total_of_special_requests', 'stays_in_weekend_nights', 'stays_in_week_nights']]

# Create the pair plot
sn.pairplot(subset, diag_kind='kde', palette='husl')
plt.suptitle('Pair Plot of Selected Numerical Features', y=1.02)
plt.show()
````

##### 1. Why did you pick the specific chart?

The pair plot was chosen to visualize the relationships between multiple numerical variables in the dataset. This type of chart provides a comprehensive view of the pairwise relationships between variables, including both scatter plots for the relationships and histograms or KDE plots for the distributions. It is particularly useful for identifying patterns, correlations, and potential outliers across multiple dimensions simultaneously.

##### 2. What is/are the insight(s) found from the chart?

Correlations: Identify any strong positive or negative correlations between variables, such as a potential correlation between lead time and average daily rate (ADR) or between the number of weekend and weekday stays.
